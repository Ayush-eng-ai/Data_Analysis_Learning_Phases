## Basic Probability

## Question.2: Complement Rule


```python
# Probability of rain
prob_rain = 0.4

# Probability of not raining using the complement rule
prob_not_rain = 1 - prob_rain

print(f"Probability of rain: {prob_rain}")
print(f"Probability of not raining: {prob_not_rain:.1f}")
```

## Question.3: Mutually Exclusive Events

```python
# Sample space for a dice roll
sample_space = {1, 2, 3, 4, 5, 6}

# Event A: Even numbers
event_a_even = {2, 4, 6}
# Event B: Odd numbers
event_b_odd = {1, 3, 5}

# Check if mutually exclusive (no common elements)
are_mutually_exclusive = event_a_even.isdisjoint(event_b_odd) # isdisjoint() checks if two sets have no elements in common

print(f"Are 'even number' and 'odd number' events mutually exclusive? {are_mutually_exclusive}")

# Probabilities
prob_a = len(event_a_even) / len(sample_space)
prob_b = len(event_b_odd) / len(sample_space)

# Combined probability for mutually exclusive events
combined_prob = prob_a + prob_b

print(f"Probability of an even number: {prob_a:.1f}")
print(f"Probability of an odd number: {prob_b:.1f}")
print(f"Combined probability (even or odd): {combined_prob:.1f}")
```

## Question.4:Independent Event
    If the probability of getting heads in a coin toss is 0.5 and the probability of getting 6 in a dice roll is 1/6, then what is the probability of both heads and 6 coming together?


```python
# Probability of getting Heads on a coin toss
prob_heads = 0.5

# Probability of getting 6 on a dice roll
prob_six = 1/6

# Calculate the probability of both events happening (since they are independent)
prob_heads_and_six = prob_heads * prob_six

print(f"Probability of Heads: {prob_heads}")
print(f"Probability of 6 on a dice: {prob_six:.4f}")
print(f"Probability of Heads AND 6: {prob_heads_and_six:.4f}")
```

## Question.5: Combinations
    There are 15 players in a cricket team. The captain has to choose the playing 11 players. In how many different ways can the team be selected? (Order does not matter).
    

```python
import math

def combinations(n, k):
    return math.factorial(n) / (math.factorial(k) * math.factorial(n - k))

n_players = 15
k_to_choose = 11

num_combinations = combinations(n_players, k_to_choose)

print(f"Total players: {n_players}")
print(f"Players to choose for playing 11: {k_to_choose}")
print(f"Number of ways to choose the team (combinations): {int(num_combinations)}") # int() because combinations must be whole number
```

## Question.6: Permutations
    In how many different ways can 4 different books be arranged on a shelf?

```python
import math

def permutations(n, k):
    return math.factorial(n) / math.factorial(n - k)

num_books = 4
num_to_arrange = 4 # Arranging all 4 books

num_permutations = permutations(num_books, num_to_arrange)

print(f"Total books: {num_books}")
print(f"Number of ways to arrange the books (permutations): {int(num_permutations)}")
```

## Question.7:  Non-Mutually Exclusive Events
    There are 30 students in a class. 15 students like Maths, 10 like Science, and 5 like both. What is the probability that a randomly selected student likes Maths or Science?
 

```python
total_students = 30
likes_math = 15
likes_science = 10
likes_both = 5

# Calculate individual probabilities
prob_math = likes_math / total_students
prob_science = likes_science / total_students
prob_both = likes_both / total_students

# Calculate probability of liking Math OR Science
prob_math_or_science = prob_math + prob_science - prob_both

print(f"Probability of liking Maths: {prob_math:.2f}")
print(f"Probability of liking Science: {prob_science:.2f}")
print(f"Probability of liking Both: {prob_both:.2f}")
print(f"Probability of liking Maths OR Science: {prob_math_or_science:.2f}")
```

## Question.8: Dependent Event
    You draw two cards without replacement from a deck (52 cards). What is the probability that both cards are kings?


```python
# Total cards in a deck
total_cards = 52
# Total Kings in a deck
total_kings = 4

# Probability of first card being a King
prob_first_king = total_kings / total_cards

# After drawing one King, remaining cards and Kings
remaining_cards = total_cards - 1
remaining_kings = total_kings - 1

# Probability of second card being a King GIVEN the first was a King
prob_second_king_given_first = remaining_kings / remaining_cards

# Probability of both cards being Kings
prob_both_kings = prob_first_king * prob_second_king_given_first

print(f"Probability of 1st card being King: {prob_first_king:.4f}")
print(f"Probability of 2nd card being King (given 1st was King): {prob_second_king_given_first:.4f}")
print(f"Probability of both cards being Kings (without replacement): {prob_both_kings:.6f}")
```

## Question.9:Mixed Concepts (Combinations + Probability)
    A box contains 6 red, 4 blue, and 2 green marbles. If you draw 3 marbles at random (without replacement), what is the probability that you will get 1 red, 1 blue, and 1 green marble?


```python
import math

def combinations(n, k):
    return math.factorial(n) / (math.factorial(k) * math.factorial(n - k))

# Define the number of marbles
red_marbles = 6
blue_marbles = 4
green_marbles = 2
total_marbles = red_marbles + blue_marbles + green_marbles

# 1. Calculate Total Possible Ways to choose 3 marbles from 12
total_ways_to_choose_3 = combinations(total_marbles, 3)
print(f"Total possible ways to choose 3 marbles: {int(total_ways_to_choose_3)}")

# 2. Calculate Favorable Ways to choose 1 Red, 1 Blue, and 1 Green
ways_to_choose_1_red = combinations(red_marbles, 1)
ways_to_choose_1_blue = combinations(blue_marbles, 1)
ways_to_choose_1_green = combinations(green_marbles, 1)

favorable_ways = ways_to_choose_1_red * ways_to_choose_1_blue * ways_to_choose_1_green
print(f"Favorable ways (1 Red, 1 Blue, 1 Green): {int(favorable_ways)}")

# 3. Calculate Probability
probability = favorable_ways / total_ways_to_choose_3
print(f"Probability of getting 1 Red, 1 Blue, 1 Green: {probability:.4f}")
```
